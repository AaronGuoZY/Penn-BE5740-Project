```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Define the directory path
directory = '/home/aarongzy/class/src/data'

# Get a list of all the npz files in the directory
file_list = [file for file in os.listdir(directory) if file.endswith('.npz')]

# Initialize counter
counter = 0

# Iterate over each file
for file_name in file_list:
    # Check if counter reaches 20
    if counter == 5:
        break
    
    # Increment counter
    counter += 1
    
    # Construct the file path
    file_path_npz = os.path.join(directory, file_name)
    
    # Load the npz file
    data = np.load(file_path_npz)
    
    # Get the list of matrix names in the npz file
    matrix_names = data.files
    
    # Iterate over each matrix
    for matrix_name in matrix_names:
        matrix_data = data['xi']  # Modify this line to access the 'xi' matrix
        max_intensity = np.max(matrix_data)
        min_intensity = np.min(matrix_data)

        plt.imshow(matrix_data, cmap='gray')
        plt.title("Matrix: xi")  # Modify the title accordingly
        plt.colorbar()
        plt.show()

        print(f"File: {file_name}")
        print("Matrix: xi")  # Modify the matrix name accordingly
        print(f"Max Intensity: {max_intensity}")
        print(f"Min Intensity: {min_intensity}")
        print()
        
    

```
